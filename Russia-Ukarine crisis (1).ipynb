```python
#Importing neccesary libraries to perform our funtions.
import pandas as pd
import numpy as np
#Now we need to lead our dataframe named personnel loss
load_personnel_loss=pd.read_csv("russia_losses_personnel.csv")
print(load_personnel_loss)
```

```python
#Now lets get a quick insight into the dataframe
load_personnel_loss.head()
```

```python
load_personnel_loss.tail()
```

```python
#As we can see that the column "personnel*" doesnt add any value to our insight we can drop it
load_personnel_loss.drop(columns =['personnel*'],inplace=True)
load_personnel_loss
```

```python
#Now let us understand the trend of how the situation in war is unfurling in terms of death toll and prisoners of war


# importing the required module
import matplotlib.pyplot as plt
 
# x axis values
x = load_personnel_loss["day"]
# corresponding y axis values
y = load_personnel_loss["personnel"]
 
# plotting the points
plt.plot(x, y)
 
# naming the x axis
plt.xlabel('As days increased')
# naming the y axis
plt.ylabel('Personnel loss')
 
# giving a title to my graph
plt.title('Personnel loss nearly 1.5 month after the war begun')
 
# function to show the plot
plt.show()
```

```python
#Repeating the same for Prisoners of War

# importing the required module
import matplotlib.pyplot as plt
 
# x axis values
x = load_personnel_loss["day"]
# corresponding y axis values
y = load_personnel_loss["POW"]
 
# plotting the points
plt.plot(x, y)
 
# naming the x axis
plt.xlabel('As days increased')
# naming the y axis
plt.ylabel('POWs captured')
 
# giving a title to my graph
plt.title('POW captured over 1.5 month')
 
# function to show the plot
plt.show()
```

```python
#Analysing the 2 graphs

#Personnel loss
#We can see that rate of personnel loss increased almost linearly as the days passed by suggesting fair further mortality in future.

#POW capture
#Interestingly though personnel loss increased linearly, POW captured rise with a linear function till about initial 18 days.
#Thereafter, number of POW captures stagnates with negligible growth.

#This may indicate a complicated scenario. Since POWs are a nations respomsibility which must be returned as per Interntional
#conventions,the trend of stagnant POW captured and rising personnel mortality suggests deteriorating human rights situations.
#and violation of principles of Universal Declaration of Human Rights,1948.
```
